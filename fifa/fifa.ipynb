```python
df = pd.DataFrame(data_frame, columns = ['Name', 'Wage', 'Value'])
df
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
```

```python
TOOLTIPS = [
    ('index', '$index'),
    ('(Wage, Value)', '(@Wage, @Value)'),
    ('Name', '@Name')
]

hover = HoverTool(tooltips = TOOLTIPS)

# create a new plot
p = figure(title = 'Fifa', x_axis_label = 'Wage', y_axis_label = 'Value', plot_width = 700, plot_height = 700, tools=[hover])

# add a circle renderer with legend
p.circle('Wage', 'Value', legend_label = 'Player', size = 10, source = df)

# show the results
show(p)
```
