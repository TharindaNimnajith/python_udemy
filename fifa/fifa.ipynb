```python
# Dropping multiple columns
data_frame = data_frame.drop(['Flag', 'Photo'], axis = 1)
data_frame.head()
```

```python
# Map feature values
# data_frame['Gender'] = data_frame['Gender'].map({'Male': 0, 'Female': 1}).astype(int)
# data_frame['Nationality'] = data_frame['Nationality'].map({'Argentina': 0, 'Portugal': 1}).astype(int)
data_frame['Nationality'] = data_frame['Nationality'].map({'Argentina': 0, 'Portugal': 1}).astype(str)
data_frame.head()
```

```python
# Lambda expressions to create new features

def get_family_size(sibsp, parch):
    family_size = sibsp + parch
    return family_size

data_frame['gk'] = data_frame[['GKDiving', 'GKHandling']].apply(lambda x: get_family_size(x['GKDiving'], x['GKHandling']), axis = 1)
data_frame.head()
```

```python
# Reorganize headers
data_frame = data_frame[['Name', 'Age', 'Overall']]
data_frame.head()
```

```python
# Saving dataframe to CSV
data_frame.to_csv('processed_data.csv', index = False)
```
